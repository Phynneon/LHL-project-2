# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import numpy as np
import pandas as pd
import sqlite3 as sq3
import requests
import os
```

```python
#Setting up the url for the api
url = 'http://api.citybik.es/v2/networks/avelo-quebec'
result = requests.get(url)
```

```python
json_data = result.json()
```

```python
#Settin up a dataframe for all the data from the API request
bike_station = pd.DataFrame(json_data['network']['stations'])
type(bike_station)
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
#Selecting our data of interest and making a new columns for total number of bikes in a station.
bike_data = bike_station[['name','empty_slots','free_bikes','latitude','longitude']]
bike_data['total_bike'] = bike_data['empty_slots'] + bike_data['free_bikes']
```

Put your parsed results into a DataFrame.
